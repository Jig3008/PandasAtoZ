```python
#We normally create a dataframe by using the DataFrame function of Pandas

#Example 1

df1 = pd.DataFrame({
    'Row ID' : [1,2,3],
    'Name' : ['A','B','C'],
    'Role' : ['Engg','Artist','Gamer'],
    'Location' : ['X','Y','Z']
})

print(df1)
```

```python
#Reading a file and then getting its top 10 rows plus its info

imbdMovieDf = pd.read_csv('IMDBMovies.csv')
imbdMovieDf.head(10)
```

```python
imbdMovieDf.info()
```

```python
'''
If we want to get the entire information from a specific column
we can write either df[column_name] or df.column_name (this would work
in case the column follows all column naming standards)
'''
movie_name = imbdMovieDf['name']
print(movie_name)
print(type(movie_name)) #This would print the type as Series
print(type(imbdMovieDf)) #This would print the type as DataFrame
```

```python
'''
If we want to get information from multiple columns,
we would send a list of all column names
'''

movie_name_year = imbdMovieDf[['name','year']]
print(movie_name_year)
print(type(movie_name_year)) #This would print the type as DataFrame
```

```python
#If we want to get a information about a specific row, we use the iloc command

imbdMovieDf.head(10)
```

```python
row3 = imbdMovieDf.iloc[3]
print(row3)
print(type(row3))
```

```python
#If we want to print multiple rows, we can pass the series of the row numbers
#dataFrame.iloc[10:] -> 10 to the end of dataframe
#dataFrame.iloc[10:15] -> 10 to 14 of dataframe
#dataFrame.iloc[-5:] -> 5 rows from the end of dataframe

print(imbdMovieDf.iloc[-10:])
```

```python
#Lets say if we want to print the information for the movie released in the year 2008

movie_2008 = imbdMovieDf[imbdMovieDf['year'] == 2008]
print(movie_2008)
print(type(movie_2008)) #This would also retuen type as dataframe
```

```python
#Lets say if we want to print the information for the movie released in the year 2008 or year 2009

movie_2008_2009 = imbdMovieDf[(imbdMovieDf['year'] == 2008) | (imbdMovieDf['year'] == 2009)]
print(movie_2008_2009)

#This can be achieved via another command - isin()

movie_2008_2009_withIsIN = imbdMovieDf[imbdMovieDf['year'].isin(
    [2008,2009]
)]

print(movie_2008_2009_withIsIN)
```

```python
#When we want to reset the index and also have re-ordering done, we use reset_index()
df = pd.DataFrame([
  ['January', 100, 100, 23, 100],
  ['February', 51, 45, 145, 45],
  ['March', 81, 96, 65, 96],
  ['April', 80, 80, 54, 180],
  ['May', 51, 54, 54, 154],
  ['June', 112, 109, 79, 129]],
  columns=['month', 'clinic_east',
           'clinic_north', 'clinic_south',
           'clinic_west']
)

df2 = df.loc[[1, 3, 5]]
print(df2)

df3 = df2.reset_index()
print(df3)

df2.reset_index(inplace= True ,drop = True)
print(df2)
```

```python
#Exercise 1: Analysis of the ShoeFly database

orders = pd.read_csv('shoefly.csv')

#Inspect the first 5 lines of data
print(orders.head())

#Your marketing department wants to send out an email blast to everyone who ordered shoes!

emails = orders['email']
print(emails)

#Frances Palmer claims that her order was wrong. What did Frances Palmer order?

frances_palmer = orders[(orders['first_name'] == 'Frances') | (orders['last_name'] == 'Palmer')]
print(frances_palmer)

#We need some customer reviews for our comfortable shoes. Select all orders for shoe_type: clogs, boots, and ballet flats and save them to the variable comfy_shoes.
comfy_shoes = orders[orders['shoe_type'].isin(['clogs','boots','ballet flats'])]
print(comfy_shoes)
```

```python
#Some practice on Lambda function

lambdaFunction = lambda x : (x * 2) + 3
print(f'The value for function 1:{lambdaFunction(5)}')

lambdafunction2 = lambda x: x.lower()
print(f'The value for function 2:{lambdafunction2('LOWER TEXT')}')

#Create a lambda function mylambda that returns the first and last letters of a string, assuming the string is at least 2 characters long.

lambdafunction3 = lambda x : x[0] + x[-1]
print(f'The value of function 3:{lambdafunction3('This is Pandas Practice')}')

#More Complex function

def myfunction(x):
  if x > 40:
    return 40 + (x - 40) * 1.50
  else:
    return x

#The above can be written as

lambdafunction4 = lambda x: 40 + (x - 40) * 1.50 if x > 40 else x
print(f'The value for function 4:{lambdafunction4(5)}')
print(f'The value for function 4:{lambdafunction4(50)}')

#The syntax is lambda x: TRUE STATEMENT - IF CONDITION - ELSE CONDITION

#You are managing the webpage of a somewhat violent video game and you want to check that each users age is 13 or greater when they visit the site.
#Write a lambda function that takes an inputted age and either returns Welcome to BattleCity! if the user is 13 or older or You must be 13 or older
#if they are younger than 13. Your lambda function should be called mylambda.

mylambda = lambda x : 'Welcome to BattleCity!' if x >= 13 else 'You must be 13 or older'
print(mylambda(int(input('Enter the age:'))))
```
