# **Project on Aggregates:**

A/B Testing for ShoeFly.com
Our favorite online shoe store, ShoeFly.com is performing an A/B Test. They have two different versions of an ad, which they have placed in emails, as well as in banner ads on Facebook, Twitter, and Google. They want to know how the two ads are performing on each of the different platforms on each day of the week. Help them analyze the data using aggregate measures.

```python
ad_clicks = pd.read_csv('ad_clicks.csv')

#Examine the first few rows of ad_clicks.
print(ad_clicks.head())

#Your manager wants to know which ad platform is getting you the most views.
#How many views (i.e., rows of the table) came from each utm_source?

views_from_each_source = ad_clicks.groupby('utm_source').user_id.count().reset_index()
print(views_from_each_source)

#Below line is written to verify if all the sources are covered
print(ad_clicks['utm_source'].unique())

#If the column ad_click_timestamp is not null, then someone actually clicked on the ad that was displayed.
# Create a new column called is_click, which is True if ad_click_timestamp is not null and False otherwise.

ad_clicks['is_click'] = ad_clicks.ad_click_timestamp.notnull()

#We want to know the percent of people who clicked on ads from each utm_source.
#Start by grouping by utm_source and is_click and counting the number of user_id‘s in each of those groups. Save your answer to the variable clicks_by_source.

clicks_by_source = ad_clicks.groupby(['utm_source','is_click']).user_id.count().reset_index()

#Now let’s pivot the data so that the columns are is_click (either True or False), the index is utm_source, and the values are user_id.Save your results to the variable clicks_pivot.

clicks_pivot = clicks_by_source.pivot(
  columns = 'is_click',
  index = 'utm_source',
  values = 'user_id'
).reset_index()

#Create a new column in clicks_pivot called percent_clicked which is equal to the percent of users who clicked on the ad from each utm_source.Was there a difference in click rates for each source?

clicks_pivot['percent_clicked'] = \
   clicks_pivot[True] / \
   (clicks_pivot[True] +
    clicks_pivot[False])

print(clicks_pivot)

#The column experimental_group tells us whether the user was shown Ad A or Ad B. Were approximately the same number of people shown both ads?

experimental_group_count = ad_clicks.groupby('experimental_group').user_id.count().reset_index()
print(experimental_group_count)

#Using the column is_click that we defined earlier, check to see if a greater percentage of users clicked on Ad A or Ad B.

user_ad_click_count = ad_clicks.groupby(['experimental_group','is_click']).user_id.count().reset_index()
user_ad_click_count_pivot = user_ad_click_count.pivot(
  columns = 'is_click',
  index = 'experimental_group',
  values = 'user_id'
).reset_index()
print(user_ad_click_count_pivot)

#The Product Manager for the A/B test thinks that the clicks might have changed by day of the week.
#Start by creating two DataFrames: a_clicks and b_clicks, which contain only the results for A group and B group, respectively.

a_clicks = ad_clicks[ad_clicks['experimental_group'] == 'A']
b_clicks = ad_clicks[ad_clicks['experimental_group'] == 'B']

print(a_clicks.head())
print(b_clicks.head())

#For each group (a_clicks and b_clicks), calculate the percent of users who clicked on the ad by day.

a_clicks_percent = a_clicks.groupby(['is_click','day']).user_id.count().reset_index()
a_clicks_percent_pivot = a_clicks_percent.pivot(
  columns = 'day',
  index = 'is_click',
  values = 'user_id'
).reset_index()
print(a_clicks_percent_pivot)

b_clicks_percent = b_clicks.groupby(['is_click','day']).user_id.count().reset_index()
b_clicks_percent_pivot = b_clicks_percent.pivot(
  columns = 'day',
  index = 'is_click',
  values = 'user_id'
).reset_index()
print(b_clicks_percent_pivot)
```

```python
#Now lets imagine if we have two tables with no common Columns except for ID.
#When we would do a merge it would create bad data.So we use the below

orders_merge = pd.read_csv('orders_merge.csv')
print(orders_merge)
products_merge = pd.read_csv('products_merge.csv')
print(products_merge)
```

```python
orders_products = pd.merge(orders_merge, products_merge.rename(columns={'id' : 'product_id'}))
orders_products
```

```python
'''
If we don’t want to do that, we have another option. We could use the keywords left_on and right_on to specify
which columns we want to perform the merge on. In the example below, the “left” table is the one that comes
first (orders), and the “right” table is the one that comes second (customers).
This syntax says that we should match the customer_id from orders to the id in customers.

Syntax:

pd.merge(
    orders,
    customers,
    left_on='customer_id',
    right_on='id',
    suffixes=['_order', '_customer']
)
'''

orders_products = pd.merge(orders_merge,
                           products_merge,
                           left_on='product_id',
                           right_on='id',
                           suffixes=['_order','_products'])
orders_products
```

```python
#So all the examples we did until now is for the INNER MERGE, it is like an intersection
#So if there is no commanlity between two tables for some rows, those rows would not be VISIBLE
#in the new merged DF

#Hence the new concept of OUTER MERGE

'''
There are two hardware stores in town: Store A and Store B. Store A’s inventory
is in DataFrame store_a and Store B’s inventory is in DataFrame store_b.
They have decided to merge into one big Super Store! We need to merge their data
'''

store_a = pd.read_csv('store_a.csv')
store_b = pd.read_csv('store_b.csv')

store_a_b_outer = pd.merge(store_a,store_b,how='outer')
store_a_b_outer
```

```python
import pandas as pd

visits = pd.read_csv('visits.csv',
                        parse_dates=[1])
checkouts = pd.read_csv('checkouts.csv',
                        parse_dates=[1])

'''
.
Cool T-Shirts Inc. just created a website for ordering their products. They want you to analyze two datasets for them:

visits contains information on all visits to their landing page
checkouts contains all users who began to checkout on their website
'''
print(visits)
print(checkouts)

'''
We want to know the amount of time from a user\’s initial visit to the website to when they start to check out.

Use merge to combine visits and checkouts and save it to the variable v_to_c.
'''
v_to_c = pd.merge(visits, checkouts,how='inner')
print(v_to_c)

'''
In order to calculate the time between visiting and checking out, define a column of v_to_c called time that calculates the difference between checkout_time and visit_time for every row.
'''

v_to_c['time'] = v_to_c['checkout_time'] - v_to_c['visit_time']
print(v_to_c)

#Use .mean() to calculate the average time to checkout and print that value to the terminal.
print(v_to_c['time'].mean())
```

```python
import pandas as pd

visits = pd.read_csv('visits_project.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])

#Step 1: Inspect the DataFrames using print and head:
print(visits.head())
print(cart.head())
print(checkout.head())
print(purchase.head())

#Step 2: Combine visits and cart using a left merge.
visits_cart = pd.merge(visits,cart,how='left')
print(visits_cart.head())
#Step 3: How long is your merged DataFrame?
total_visits = len(visits_cart)
print(total_visits)

#Step 4: How many of the timestamps are null for the column cart_time? What do these null rows mean?
didnt_reach_cart = len(visits_cart[visits_cart['cart_time'].isnull()])
print(didnt_reach_cart)
#Step 5: What percent of users who visited Cool T-Shirts Inc. ended up not placing a t-shirt in their cart?
no_order_placed = float(didnt_reach_cart)/float(total_visits)
print(no_order_placed)

#Step 6: Repeat the left merge for cart and checkout and count null values. What percentage of users put items in their cart, but did not proceed to checkout?

cart_checkout = pd.merge(cart,checkout,how='left')
total_carts_created = len(cart_checkout)
didnt_reach_checkout = len(cart_checkout[cart_checkout['checkout_time'].isnull()])

print(total_carts_created)
print(didnt_reach_checkout)
total_no_checkout = float(didnt_reach_checkout)/float(total_carts_created)
print(total_no_checkout)

#Step 7: Merge all four steps of the funnel, in order, using a series of left merges. Save the results to the variable all_data.

all_data = visits.merge(cart,how='left').merge(checkout,how='left').merge(purchase,how='left')
print(all_data.head())

#Step 8: What percentage of users proceeded to checkout, but did not purchase a t-shirt?
reached_checkout = all_data[~all_data['checkout_time'].isnull()]
didnt_purchase = all_data[(all_data['purchase_time'].isnull()) & (~all_data['checkout_time'].isnull())]
checked_out_didnt_purchase = float(len(didnt_purchase))/float(len(reached_checkout))
print("% of users who got to checkout but did not purchase:",checked_out_didnt_purchase)

#Step 9: Which step of the funnel is weakest (i.e., has the highest percentage of users not completing it)? How might Cool T-Shirts Inc. change their website to fix this problem?
print("{} percent of users who visited the page did not add a t-shirt to their cart".format(round(no_order_placed*100, 2)))
print("{} percent of users who added a t-shirt to their cart did not checkout".format(round(total_no_checkout*100, 2)))
print("{} percent of users who made it to checkout  did not purchase a shirt".format(round( checked_out_didnt_purchase*100, 2)))

#Step 10: Using the giant merged DataFrame all_data that you created, let’s calculate the average time from initial visit to final purchase. Add a column that is the difference between purchase_time and visit_time.

all_data['total_time'] = all_data['purchase_time'] - all_data['visit_time']
print(all_data.head())

#Step 11: Examine the results by printing the new column to the screen.
print(all_data['total_time'])

#Step 12: Calculate the average time to purchase by applying the .mean() function to your new column.
print(all_data['total_time'].mean())
```
